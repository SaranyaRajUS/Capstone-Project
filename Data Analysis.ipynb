```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif
```

```python
X = dataset[['attention_span_score', 'social_skills_score', 'cognitive_test_score', 'parental_involvement_score']]
vif_df = calc_vif(X)
print(vif_df)
```

```python
from scipy.stats import pearsonr

# Drop NaNs
dataset = dataset.dropna(subset=['parental_involvement_score', 'high_IGCSE_performance'])

# Pearson correlation
corr, pval = pearsonr(dataset['parental_involvement_score'], dataset['high_IGCSE_performance'])
print(f"Correlation: {corr}, p-value: {pval}")
```

```python
from scipy.stats import chi2_contingency, ttest_ind

# Hypothesis 1: Is there a significant association between gender and IGCSE performance?
contingency_gender_perf = pd.crosstab(df['gender'], df['high_IGCSE_performance'])
chi2_gender, p_gender, _, _ = chi2_contingency(contingency_gender_perf)

# Hypothesis 2: Is there a significant difference in cognitive scores between high and low performers?
high_perf = df[df['high_IGCSE_performance'] == 1]['cognitive_test_score']
low_perf = df[df['high_IGCSE_performance'] == 0]['cognitive_test_score']
t_stat_cog, p_val_cog = ttest_ind(high_perf, low_perf, equal_var=False)

# Hypothesis 3: Is there a significant association between learning difficulty and IGCSE performance?
contingency_diff_perf = pd.crosstab(df['has_learning_difficulty'], df['high_IGCSE_performance'])
chi2_diff, p_diff, _, _ = chi2_contingency(contingency_diff_perf)

# Output the results
{
    "Gender vs Performance (Chi-square)": {"Chi2": chi2_gender, "p-value": p_gender},
    "Cognitive Score vs Performance (T-test)": {"T-stat": t_stat_cog, "p-value": p_val_cog},
    "Learning Difficulty vs Performance (Chi-square)": {"Chi2": chi2_diff, "p-value": p_diff}
}
```

```python
from scipy.stats import friedmanchisquare

# Drop missing values in relevant columns
dataset = dataset.dropna(subset=['attention_span_score', 'social_skills_score', 'cognitive_test_score'])

# Apply Friedman test
result = friedmanchisquare(
    dataset['attention_span_score'],
    dataset['social_skills_score'],
    dataset['cognitive_test_score']
)
```
