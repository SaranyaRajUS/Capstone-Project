```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dataset = pd.read_csv("student_igcse_dataset.csv")
dataset
```

```python
null_counts = dataset.isnull().sum()
dataset
```

```python
na_counts = dataset.isna().sum()
dataset
```

```python
def quanqual(dataset):
    quan = []
    qual = []
    for columnName in dataset.columns:
        if dataset[columnName].dtype == 'O':
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan, qual
```

```python
quan,qual=quanqual(dataset)
```

```python
dataset[quan]
```

```python
dataset[qual]
```

```python
imp_num = SimpleImputer(missing_values=np.nan, strategy='mean')
dataset[quan] = imp_num.fit_transform(dataset[quan])
dataset[quan]
```

```python
imp_cat = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
dataset[qual] = imp_cat.fit_transform(dataset[qual])
dataset[qual]
```

```python
label_encoders = {}
for col in qual:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col])
    label_encoders[col] = le
```

```python
scaler = StandardScaler()
dataset[quan] = scaler.fit_transform(dataset[quan])
dataset[quan]
```

```python
preprocessed = pd.concat([dataset[quan], dataset[qual]], axis=1)
```
